```python
%matplotlib inline
import numpy as np
import sigpy as sp
import sigpy.mri as mr
import sigpy.plot as pl
import scipy.io as sio
```

```python
ksp = np.array(sio.loadmat("ismrm_sunrise_cause_sim_data.mat")['im1'])
mps = np.array(sio.loadmat("ismrm_sunrise_cause_sim_data.mat")['smaps'])
print(ksp.shape, mps.shape)
```

```python
pl.ImagePlot(mps.transpose(2, 0, 1), z=0, title='Sensitivity Maps Estimated by ESPIRiT')
```

```python
im = mr.app.SenseRecon(ksp, mps, lamda=1, weights=0.5).run()
```
